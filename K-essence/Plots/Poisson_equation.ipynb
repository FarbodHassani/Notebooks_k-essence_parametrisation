```python
As=2.215*1.e-9;
h=0.67556;
kp=0.05/h; 
ns=0.9619;
c=2.99792458*1.e5; #Km/s
H0=100*h/c; #H0 in 1/Mpc unit. 0.00022593979933110373
# print("H0[1/Mpc]=100h/c: ",H0);
Omega_b=0.022032/h/h; Omega_cdm=0.12038/h/h;
Omega_m0=Omega_b+Omega_cdm;
def Hubble_conf_Mpc_lcdm(a):
    c=2.99792458*1.e5;h=0.67556;H0=100*h/c;w=-0.9;
    Omega_b=0.022032/h/h; Omega_cdm=0.12038/h/h;
    Omega_m=Omega_b+Omega_cdm; 
    Omega_rad=9.1671353942930788e-05; Omega_L=1.-Omega_m-Omega_rad;
    return H0*np.sqrt(Omega_m*(a**-3)+Omega_rad*(a**-4)+Omega_L)*a

plt.figure(figsize=(18,8))
ax = plt.gca()
ax.tick_params(axis = 'both', which = 'major', labelsize = 16)
ax.tick_params(axis = 'both', which = 'minor', labelsize = 16)
plt.figure(1)
# Newtonian_Poisson_test=[]
for i in range (1,maxNum_lcdm):
    #Making |Phi|=sqrt(P_phi), |delta| = sqrt(P_delta)
    k2=(Gev_lcdm_phi_zall[i][:,0]*h)**2;
    phi=np.sqrt(Gev_lcdm_phi_zall[i][:,1]);
    delta=np.sqrt(Gev_lcdm_deltam_zall[i][:,1]);
    Newtonian_Poisson_diff= np.abs(k2*phi -(3./2.)*Hubble_conf_Mpc_lcdm(1.)**2 * (Omega_m0*(1.+z_lcdm[i])*delta) )
    Newtonian_Poisson_rel=Newtonian_Poisson_diff/(k2*phi);
    plt.loglog(Gev_lcdm_phi_zall[i][:,0], Newtonian_Poisson_rel[:],color=ColorsI[i],linestyle='solid',lw=2.5,  label=r" z="+str(z_lcdm[i]))
    #     plt.loglog(class_phi_kess_zall[i][:,0], class_phi_kess_zall[i][:,0]**3 * class_delta_kess_zall[i][:,1]/(2. * np.pi**2),color=ColorsI[i],linestyle='dashed',lw=2.5,  label=r"class, z="+str(z_list[i]))
#     plt.loglog(Gev_fld_deltam_zall[i][:,0], Gev_fld_deltam_zall[i][:,1],color=ColorsI[i+1],linestyle='dashed',lw=2.5,  label=r"realized class, z="+str(z_list[i]))
plt.ylim(1e-1,4.81e-1)
# plt.ylim(1e-1,4.1)
# plt.yticks(np.arange(4.68e-1,4.8e-1,0.002))
# plt.ylim(4.71e-2,4.81e-2)

plt.xlim(2.e-2,2.e0)
plt.legend(bbox_to_anchor=(0.1, 0.80, 0.3, .102), loc=1,ncol=1,fontsize=12, mode="expand", borderaxespad=0.)
plt.xlabel("k[h/Mpc]",fontsize=18)
plt.ylabel(r"$|\frac{k^2 \sqrt{\mathcal{P}_{\Phi}}  - ( 3/2) \mathcal{H}_0^2 \frac{\Omega_m (0) }{a} \sqrt{\mathcal{P}_{\delta}}} {k^2 \sqrt{\mathcal{P}_{\Phi}}}|$ ",fontsize=18)
plt.show()
```

```python
As=2.215*1.e-9;
h=0.67556;
kp=0.05/h; 
ns=0.9619;
c=2.99792458*1.e5; #Km/s
H0=100*h/c; #H0 in 1/Mpc unit. 0.00022593979933110373
# print("H0[1/Mpc]=100h/c: ",H0);
Omega_b=0.022032/h/h; Omega_cdm=0.12038/h/h;
Omega_m0=Omega_b+Omega_cdm;
def Hubble_conf_Mpc_lcdm(a):
    c=2.99792458*1.e5;h=0.67556;H0=100*h/c;w=-0.9;
    Omega_b=0.022032/h/h; Omega_cdm=0.12038/h/h;
    Omega_m=Omega_b+Omega_cdm; Omega_Lambda=0;
    Omega_rad=9.1671353942930788e-05; Omega_L=1.-Omega_m-Omega_rad;
    return H0*np.sqrt(Omega_m*(a**-3)+Omega_rad*(a**-4)+Omega_L)*a

plt.figure(figsize=(18,8))
ax = plt.gca()
ax.tick_params(axis = 'both', which = 'major', labelsize = 16)
ax.tick_params(axis = 'both', which = 'minor', labelsize = 16)
plt.figure(1)
# Newtonian_Poisson_test=[]
for i in range (1,maxNum_lcdm):
    #Making |Phi|=sqrt(P_phi), |delta| = sqrt(P_delta)
    k2=Gev_lcdm_phi_zall[i][:,0]**2;
    phi=np.sqrt(Gev_lcdm_phi_zall[i][:,1]);
    delta=np.sqrt(Gev_lcdm_deltam_zall[i][:,1]);
    Relativistic_Poisson_diff= np.abs(k2*phi + 3.*Hubble_conf_Mpc_lcdm(1./(1+z_lcdm[i]))**2 *phi   -(3./2.)*Hubble_conf_Mpc_lcdm(1.)**2 * (Omega_m0*(1.+z_lcdm[i])*delta) )
    Relativistic_Poisson_rel=Relativistic_Poisson_diff/(k2*phi);
    
    plt.loglog(Gev_lcdm_phi_zall[i][:,0], Relativistic_Poisson_rel[:],color=ColorsI[i],linestyle='solid',lw=2.5,  label=r" z="+str(z_lcdm[i]))
    #     plt.loglog(class_phi_kess_zall[i][:,0], class_phi_kess_zall[i][:,0]**3 * class_delta_kess_zall[i][:,1]/(2. * np.pi**2),color=ColorsI[i],linestyle='dashed',lw=2.5,  label=r"class, z="+str(z_list[i]))
#     plt.loglog(Gev_fld_deltam_zall[i][:,0], Gev_fld_deltam_zall[i][:,1],color=ColorsI[i+1],linestyle='dashed',lw=2.5,  label=r"realized class, z="+str(z_list[i]))
# plt.ylim(4.71e-1,4.81e-1)
# plt.ylim(1e-1,4.1)
plt.yticks(np.arange(4.68e-1,4.8e-1,0.002))
plt.ylim(4.71e-1,4.81e-1)

plt.xlim(2.e-2,2.e0)
plt.legend(bbox_to_anchor=(0.1, 0.80, 0.3, .102), loc=1,ncol=1,fontsize=12, mode="expand", borderaxespad=0.)
plt.xlabel("k[h/Mpc]",fontsize=18)
plt.ylabel(r"$|\frac{k^2 \sqrt{\mathcal{P}_{\Phi}} + 3\mathcal{H}^2\sqrt{\mathcal{P}_{\Phi}} - ( 3/2) \mathcal{H}_0^2 \frac{\Omega_m (0) }{a} \sqrt{\mathcal{P}_{\delta}}} {k^2 \sqrt{\mathcal{P}_{\Phi}}}|$ ",fontsize=18)
plt.show()
```
